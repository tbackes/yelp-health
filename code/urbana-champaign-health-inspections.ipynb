```python
import pandas as pd
import numpy as np
import requests as requests
from bs4 import BeautifulSoup, UnicodeDammit
import StringIO
import logging
import time
import pickle
```

```python
def single_query(payload, s=None):
    link = 'http://champaign.il.gegov.com/_templates/90/Food_Establishment_Inspection/_report_full.cfm'
    if s is None:
        response = requests.get(link, params=payload)
    else:
        response = s.get(link, params=payload)
    #return BeautifulSoup(response.content, from_encoding='UTF-8')
    return response.content


def merge_two_dicts(x, y):
    '''Given two dicts, merge them into a new dict as a shallow copy.'''
    z = x.copy()
    z.update(y)
    return z

def open_pickle(f_name):
    with open(f_name, 'rb') as f:
        data = pickle.load(f)
    return data


def save_to_pickle(data, fname):
    with open(fname, 'wb') as handle:
        pickle.dump(data, handle)
```
